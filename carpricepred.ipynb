```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('carpricepred.csv')
df.columns
df
```

```python
df.isnull().sum()
```

```python
# proportion of each type 

import matplotlib.pyplot as plt 
# (df['fuel_type'].value_counts() / df['fuel_type'].shape) * 100
# this line will include null values in the denominator showing the proportion of each feature wrt entire dataset 

# Will normalise it 
proportion = df['fuel_type'].value_counts(normalize=True) * 100 # Normalise works as such total_value_category / total_non_null_values
miss_val = df['fuel_type'].isnull().sum()


pet_count = int(round(proportion['Petrol']/100 * miss_val))
die_count = int(round(proportion['Diesel']/100 * miss_val))


miss_indices = df[df['fuel_type'].isnull()].index
df.loc[miss_indices[:pet_count], 'fuel_type'] = 'Petrol'
df.loc[miss_indices[pet_count:pet_count + die_count], 'fuel_type'] = 'Diesel'


```

```python
df['kms_driven']=df['kms_driven'].str.split().str.get(0).str.replace(',','')
```

```python
df['kms_driven']=pd.to_numeric(df['kms_driven'],errors='coerce') # (errors)if any non numeric type found it converts it into null value NaN
df['kms_driven']
median = df['kms_driven'].median()
df['kms_driven'] = df['kms_driven'].fillna(median)
```
